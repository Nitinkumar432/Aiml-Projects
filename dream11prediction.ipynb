# ***Merging scorecards with her match***

```python
# Rename the column 'for' to 'For' in the 'scorecards' dataset
scorecards.rename(columns={'for': 'For'}, inplace=True)
scorecards.rename(columns={'against': 'Against'}, inplace=True)
# Merge scorecards dataset with output dataset on 'For' and 'Against' columns
merged_scorecards_output = pd.merge(scorecards, output, on=['For', 'Against'], how='inner')

# Display the merged data
print(merged_scorecards_output)
```

# ***Merging scorecards with players***

```python
# Assuming the career dataset has a column 'Player' as well
# Merge career dataset with the player name
player_career = pd.merge(career, season[['Player']], on='Player', how='inner')

# Merge player_career with merged_scorecards_output on 'Player' column
final_merged_data = pd.merge(merged_scorecards_output, player_career, on='Player', how='inner')

# Display the final merged data
print(final_merged_data)
```

# ***Check PLayers Stats***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming merged_data is already defined

def display_stats(player_stats):
    total_runs = player_stats['Runs'].sum()
    highest_score = player_stats['HS'].max()
    average = player_stats['Ave'].mean()


    print("Player Statistics:")
    print(f"Total Runs: {total_runs}")
    print(f"Highest Score: {highest_score}")
    print(f"Average: {average:.2f}")


def plot_career_runs(player_stats):
    plt.figure(figsize=(12, 8))
    sns.barplot(data=player_stats, x='Mat', y='Runs')
    plt.title("Career Runs")
    plt.xlabel("Matches")
    plt.ylabel("Runs")
    plt.show()

def plot_best_innings(player_stats):
    best_innings = player_stats.nlargest(5, 'Runs')
    plt.figure(figsize=(10, 6))
    sns.barplot(data=best_innings, x='Runs', y='HS', palette='viridis')
    plt.title("Best Innings")
    plt.xlabel("Runs")
    plt.ylabel("Highest Score")
    plt.show()

def plot_best_season(player_stats):
    best_season = player_stats[player_stats['Runs'] == player_stats['Runs'].max()]
    plt.figure(figsize=(8, 6))
    sns.barplot(data=best_season, x='Season', y='Runs', palette='rocket')
    plt.title("Best Season")
    plt.xlabel("Season")
    plt.ylabel("Runs")
    plt.show()

def plot_last_10_matches(player_stats):
    last_10_matches = player_stats.tail(10)
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=last_10_matches, x='Mat', y='Runs', marker='o')
    plt.title("Performance in Last 10 Matches")
    plt.xlabel("Matches")
    plt.ylabel("Runs")
    plt.show()

def plot_season_performance(player_stats):
    plt.figure(figsize=(12, 8))
    sns.lineplot(data=player_stats, x='Season', y='Runs', marker='o')
    plt.title("Season-wise Performance")
    plt.xlabel("Season")
    plt.ylabel("Runs")
    plt.show()

def plot_player_stats(player_name):
    # Clean player name to ensure consistency
    player_name = player_name.lower()

    # Filter career dataset for the player's stats
    player_stats = merged_data[merged_data['Player'].str.lower() == player_name]

    if len(player_stats) == 0:
        print("Player not found.")
        return

    # Display player statistics
    display_stats(player_stats)

    # Plotting visualizations
    plot_career_runs(player_stats)
    plot_best_innings(player_stats)
    plot_best_season(player_stats)
    plot_last_10_matches(player_stats)
    plot_season_performance(player_stats)

# Example usage:
player_name = input("Enter player name: ")
plot_player_stats(player_name)
```

# ***TOP 5 Six Hitters***

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def best_six_hitter_in_season(season):
    # Filter data for the specified season
    season_data = merged_data[merged_data['Season'] == season]

    if season_data.empty:
        print(f"No data available for the season {season}.")
        return

    # Group by player and calculate the total number of sixes hit
    sixes_by_player = season_data.groupby('Player')['6s_y'].sum().reset_index()

    # Find the top 5 six hitters
    top_five_hitters = sixes_by_player.nlargest(5, '6s_y')

    # Get the best six hitter's data
    best_six_hitter = sixes_by_player[sixes_by_player['6s_y'] == sixes_by_player['6s_y'].max()]
    best_six_hitter_name = best_six_hitter['Player'].iloc[0]
    best_six_hitter_data = season_data[season_data['Player'] == best_six_hitter_name]

    # Plotting the graph
    plt.figure(figsize=(12, 8))

    # Plot the best six hitter
    sns.barplot(data=best_six_hitter_data, x='match', y='6s_y', color='skyblue', label=f"{best_six_hitter_name} (Best)")

    # Plot the top 5 six hitters
    for i, row in top_five_hitters.iterrows():
        if row['Player'] != best_six_hitter_name:  # Skip the best hitter
            player_data = season_data[season_data['Player'] == row['Player']]
            sns.barplot(data=player_data, x='match', y='6s_y', alpha=0.5, label=row['Player'])

    plt.title(f"Sixes Hit per Match for Top Six Hitters in Season {season}")
    plt.xlabel("Match")
    plt.ylabel("Number of Sixes")
    plt.xticks(rotation=90)
    plt.legend(title="Player", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

    # Display the top 5 six hitters' statistics in a table-like view
    top_five_hitters_stats = sixes_by_player.merge(top_five_hitters, on='Player', how='inner')
    top_five_hitters_stats = top_five_hitters_stats.rename(columns={'6s_y_x': 'Total Sixes', '6s_y_y': 'Sixes per Match'})
    print(f"\nTop 5 Six Hitters in Season {season}:\n")
    print(top_five_hitters_stats.to_string(index=False))

# Example usage:
season = int(input("Enter Season Name : "))
best_six_hitter_in_season(season)
```

# ***Top 5 Four Hitters by Season***

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def best_four_hitter_in_season(season):
    # Filter data for the specified season
    season_data = merged_data[merged_data['Season'] == season]

    if season_data.empty:
        print(f"No data available for the season {season}.")
        return

    # Group by player and calculate the total number of fours hit
    fours_by_player = season_data.groupby('Player')['4s_y'].sum().reset_index()

    # Find the top 5 four hitters
    top_five_hitters = fours_by_player.nlargest(5, '4s_y')

    # Get the best four hitter's data
    best_four_hitter = fours_by_player[fours_by_player['4s_y'] == fours_by_player['4s_y'].max()]
    best_four_hitter_name = best_four_hitter['Player'].iloc[0]
    best_four_hitter_data = season_data[season_data['Player'] == best_four_hitter_name]

    # Plotting the graph
    plt.figure(figsize=(12, 8))

    # Plot the best four hitter
    sns.barplot(data=best_four_hitter_data, x='match', y='4s_y', color='lightcoral', label=f"{best_four_hitter_name} (Best)")

    # Plot the top 5 four hitters
    for i, row in top_five_hitters.iterrows():
        if row['Player'] != best_four_hitter_name:  # Skip the best hitter
            player_data = season_data[season_data['Player'] == row['Player']]
            sns.barplot(data=player_data, x='match', y='4s_y', alpha=0.5, label=row['Player'])

    plt.title(f"Fours Hit per Match for Top Four Hitters in Season {season}")
    plt.xlabel("Match")
    plt.ylabel("Number of Fours")
    plt.xticks(rotation=90)
    plt.legend(title="Player", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

    # Display the top 5 four hitters' statistics in a table-like view
    top_five_hitters_stats = fours_by_player.merge(top_five_hitters, on='Player', how='inner')
    top_five_hitters_stats = top_five_hitters_stats.rename(columns={'4s_y_x': 'Total Fours', '4s_y_y': 'Fours per Match'})
    print(f"\nTop 5 Four Hitters in Season {season}:\n")
    print(top_five_hitters_stats.to_string(index=False))

# Example usage:
season = int(input("Enter Season Name : "))
best_four_hitter_in_season(season)
```
